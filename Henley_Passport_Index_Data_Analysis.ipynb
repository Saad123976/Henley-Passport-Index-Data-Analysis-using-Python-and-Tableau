```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
country_lists = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-09-09/country_lists.csv')
```

```python
rank_by_year = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-09-09/rank_by_year.csv')
```

```python
rank_by_year.set_index('year', inplace=True)
```

```python
## Which countries have made the most dramatic improvements in passport power?

rank_by_year.pivot_table(columns='year', index='country', values='visa_free_count')
```

```python
rank_by_year.pivot_table(columns='year', index='country', values='rank')
```

```python
## Countries with most drastic improvement in ranking (2006 - 2025)

rank_2006 = rank_by_year.loc[2006].set_index('country')['rank']
rank_2025 = rank_by_year.loc[2025].set_index('country')['rank']

rank_difference = rank_2025 - rank_2006
rank_difference.sort_values(ascending=True).head(10)
```

```python
## Countries with most drastic improvement in visa_free_count (2006 - 2025)

visa_free_count_2006 = rank_by_year.loc[2006].set_index('country')['visa_free_count']
visa_free_count_2025 = rank_by_year.loc[2025].set_index('country')['visa_free_count']

visa_free_count_difference = visa_free_count_2025 - visa_free_count_2006
```

```python
visa_free_count_difference.sort_values(ascending=False).head(10).plot(kind='bar', figsize=(15,5),color='aqua')
plt.title('Top 10 Countries with Highest Increase in Visa-Free Count (2006 - 2025)')
plt.xlabel('Country')
plt.ylabel('Change in Visa-Free Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
## Which passports have lost the most visa-free access?

visa_free_count_difference.sort_values(ascending=True).head(10).plot(kind='bar', figsize=(15,5),color='purple')
plt.title('Top 10 Countries with Decrease or Slightest Increase in Visa-Free Count (2006 - 2025)')
plt.xlabel('Country')
plt.ylabel('Change in Visa-Free Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
## Did the COVID-19 pandemic affect passport rankings, particularly for countries that implemented strict border controls? (Pre vs Post Pandemic)

rank_2019 = rank_by_year.loc[2019].set_index('country')['rank']
rank_2022 = rank_by_year.loc[2022].set_index('country')['rank']

rank_difference_covid = rank_2022 - rank_2019
```

```python
rank_difference_covid.sort_values(ascending=True).head(10).plot(kind='bar', figsize=(15,5), color='orange')
plt.title('Top 10 Countries with Largest Improvement in Rankings (Pre vs Post Pandemic)')
plt.ylabel('Change in Ranking')
plt.xticks(rotation=45, ha='right')
plt.show()

rank_difference_covid.sort_values(ascending=False).head(10).plot(kind='bar', figsize=(15,4), color='green')
plt.title('Top 10 Countries with Largest Decline in Rankings (Pre vs Post Pandemic)')
plt.ylabel('Change in Ranking')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
## Did the COVID-19 pandemic affect visa_free_access, particularly for countries that implemented strict border controls? (Pre vs Post Pandemic)

visa_free_count_2019 = rank_by_year.loc[2019].set_index('country')['visa_free_count']
visa_free_count_2022 = rank_by_year.loc[2022].set_index('country')['visa_free_count']

visa_free_count_covid = visa_free_count_2022 - visa_free_count_2019
```

```python
visa_free_count_covid.sort_values(ascending=False).head(10).plot(kind='bar', figsize=(15,5))
plt.title('Top 10 Countries with Largest Increase in Visa-Free Count (Pre vs Post Pandemic)')
plt.xlabel('Country')
plt.ylabel('Change in Visa-Free Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

visa_free_count_covid.sort_values(ascending=True).head(10).plot(kind='bar', figsize=(15,5), color='red')
plt.title('Top 10 Countries with Largest Decrease in Visa-Free Count (Pre vs Post Pandemic)')
plt.xlabel('Country')
plt.ylabel('Change in Visa-Free Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
## Do countries with stronger relationships tend to offer mutual visa-free access?

visa_free_access_countries = country_lists.set_index('country')['visa_free_access'].str.extractall(r'"name":\s*"([^"]+)"')
x = visa_free_access_countries.groupby('country').count().sort_values(ascending=False, by=0).head(10)
print(f'Countries with most visa free access:{x}')
```

```python
## How many countries require a visa versus those that don't?

visa_required_countries = country_lists.set_index('country')['visa_required'].str.extractall(r'"name":\s*"([^"]+)"')
x = visa_required_countries.groupby('country').count().sort_values(ascending=False,by=0).head(10)
print(f'Top countries which require a visa: {x}')
```

```python
# How many countries require a visa for travelers from the USA?

visa_online_countries = country_lists.set_index('country')['visa_online'].str.extractall(r'"name":\s*"([^"]+)"')
visa_online_countries.query('country == "United States"').count()
print(f"Countries that require a visa for USA travelers: {visa_online_countries.query('country == "United States"').count()}")
```

```python
# Which countries offer visa on arrival for travelers from Canada?

visa_arrival_countries = country_lists.set_index('country')['visa_on_arrival'].str.extractall(r'"name":\s*"([^"]+)"')
visa_arrival_countries.query('country == "Canada"')
print(f"Countries with visa on arrival for Canadian travelers: {visa_arrival_countries.query('country == "Canada"')}")
```

```python
# How many countries offer visa-free access to travelers from the UK?

visa_free_access_countries.query('country == "United Kingdom"').count()
print(f"Visa-free countries for UK travelers: {visa_free_access_countries.query('country == "United Kingdom"').count()}")
```

```python
# List all countries where travelers from India can apply for an eTA.

electronic_travel_authorisation_countries = country_lists.set_index('country')['electronic_travel_authorisation'].str.extractall(r'"name":\s*"([^"]+)"')
electronic_travel_authorisation_countries.query('country == "India"')
print(f"Countries where Indian travelers need an eTA: {electronic_travel_authorisation_countries.query('country == "India"')}")
```

```python
# What percentage of countries allow online visa applications?

x = round(visa_online_countries.groupby('country').count()/country_lists['country'].nunique()*100,2).sort_values(ascending=False, by=0)
print(f'Percentage of countries that allow online applications: {x}')
```

```python
# Visa Type Distribution by Country

visa_categories = [electronic_travel_authorisation_countries.count(), visa_arrival_countries.count(), visa_online_countries.count(),visa_required_countries.count(),visa_free_access_countries.count()]

plt.figure(figsize=(20, 6))
plt.bar(['electronic_travel_authorisation','visa_arrival','visa_online','visa_required','visa_free_access'],[cat[0] for cat in visa_categories], color='coral')
plt.title('Number of Countries in Each Visa Category')
plt.ylabel('Number of Countries')
plt.xlabel('Visa Categories')
plt.tight_layout()
plt.show()
```
